## Rule-based Extraction

```python
import re
```

```python
def extract_definiens(definition):
    """
    Extracts the definiens from a mathematical definition using regular expressions.
    This assumes the definition follows a pattern where 'is' or similar verbs introduce the definiens.
    """
    # Regular expression to capture the definiens after phrases like "is", "are", "refers to", etc.
    definiens_pattern = r'\b(?:is|are|refers to|means|can be defined as)\b\s(.+?)\.'
    
    # Find matches
    match = re.search(definiens_pattern, definition)
    
    if match:
        return match.group(1).strip()
    else:
        return "No definiens found."

# Example usage
definition = "A circle is a shape with all points equidistant from its center."
definiens = extract_definiens(definition)
print("Extracted definiens:")
print(definiens)
```
