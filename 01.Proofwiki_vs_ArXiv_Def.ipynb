## How many terms are duplicated both in arXiv papers and in proofwiki? 99 unique values.

```python
arxiv_duplicated_terms = [s.lower() for s in df_arxiv_duplicated_term['plain_text_term'] ]
common_def_pw = df_proofwiki_dd[ df_proofwiki_dd['term'].str.lower().isin(arxiv_duplicated_terms)]
common_def_pw
```

```python
common_set_pw = set(common_def_pw['term'].str.lower().to_list())
len(common_set_pw), common_set_pw
```

```python
pw_dd_terms = [s.lower() for s in df_proofwiki_dd['term'] ]
common_def_ax = df_arxiv_duplicated_term[ df_arxiv_duplicated_term['plain_text_term'].str.lower().isin(pw_dd_terms) ]
common_def_ax
```

```python
common_set_ax = set(common_def_ax['plain_text_term'].str.lower().to_list())
len(common_set_ax), common_set_ax
```

## How many proofwiki definienda are conatiend in paper-defined definienda

```python
len(common_def_pw['plain_text_definition'].unique())
```

```python
proofwiki_dd_terms = set(df_proofwiki_dd['term'])
searchfor_terms_str = '|'.join(set(proofwiki_dd_terms)).casefold()
mentionned_definienda = df_arxiv_duplicated_term[ df_arxiv_duplicated_term['plain_text_term'].str.contains(searchfor_terms_str, case=False) ]
len(mentionned_definienda)
```

```python
mentioned_terms = [ [t for t in proofwiki_dd_terms if t.lower() in ptt.lower()]  for ptt in mentionned_definienda['plain_text_term']]
```

```python
len(mentioned_terms), mentioned_terms
```

```python
mentionned_definienda['mentioned_proofwiki_terms'] = mentioned_terms
mentionned_definienda
```

```python
mentionned_definienda['mentioned_proofwiki_terms'].value_counts().to_csv("mentioned_proofwiki_terms_value_counts.csv")
```

```python
mentionned_definienda.to_csv("arXiv_terms_including_proofwiki_terms.csv", index=False)
```
